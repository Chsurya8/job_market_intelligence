```python

import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../data/indian-job-market-dataset-2025.xlsx")
```

```python
df = df.dropna()
```

```python

df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df["minimumSalary"] = pd.to_numeric(
    df["minimumSalary"],
    errors="coerce"
)

df["maximumSalary"] = pd.to_numeric(
    df["maximumSalary"],
    errors="coerce"
)
```

```python
df["avg_salary"] = (
    df["minimumSalary"] +
    df["maximumSalary"]
) / 2
```

```python
df["minimumExperience"] = pd.to_numeric(
    df["minimumExperience"],
    errors="coerce"
)

df["maximumExperience"] = pd.to_numeric(
    df["maximumExperience"],
    errors="coerce"
)
```

```python
df["avg_experience"] = (
    df["minimumExperience"] +
    df["maximumExperience"]
) / 2
```

```python
df["AggregateRating"] = (
    df["AggregateRating"]
    .fillna(0)
)
```

```python
df["ReviewsCount"] = (
    df["ReviewsCount"]
    .fillna(0)
)
```

```python
df["location"] = (
    df["location"]
    .str.strip()
    .str.title()
)
```

```python
df["companyName"] = (
    df["companyName"]
    .str.strip()
)
```

```python
df.to_csv(
    "cleaned_jobs.csv",
    index=False
)
```

```python
df["ReviewsCount"] = (
    pd.to_numeric(df["ReviewsCount"], errors="coerce")
    .fillna(0)
    .astype(int)
)
```

```python
df['salary_available'] = df['avg_salary'] > 0
df['experience_required'] = df['minimumExperience'] > 0
```

```python
salary_df = df[df['avg_salary'] > 0]
```

```python
salary_disclosure_rate = (
    len(df[df['avg_salary'] > 0]) / len(df)
) * 100
```

```python
df['salary_range'] = df['maximumSalary'] - df['minimumSalary']
```

```python
df['job_level'] = pd.cut(
    df['minimumExperience'],
    bins=[-1,0,2,5,100],
    labels=['Fresher','Junior','Mid','Senior']
)
```

```python
df['minimumSalary'] = df['minimumSalary'].replace(0, np.nan)
df['maximumSalary'] = df['maximumSalary'].replace(0, np.nan)
df['avg_salary'] = df['avg_salary'].replace(0, np.nan)
df['salary_range'] = df['salary_range'].replace(0, np.nan)
```

```python
df.to_csv("../data/cleaned_jobs.csv", index=False)

df.to_excel("../data/cleaned_jobs.xlsx", index=False)
```
